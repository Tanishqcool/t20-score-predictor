```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.DataFrame()
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        frames = [final_df, df]
        final_df = pd.concat(frames)
        counter += 1
        
final_df
```

```python
final_df.drop(columns=[
    'meta.data_version',
    'meta.created',
    'meta.revision',
    'info.outcome.bowl_out',
    'info.bowl_out',
    'info.supersubs.South Africa',
    'info.supersubs.New Zealand',
    'info.outcome.eliminator',
    'info.outcome.result',
    'info.outcome.method',
    'info.neutral_venue',
    'info.match_type_number',
    'info.outcome.by.runs',
    'info.outcome.by.wickets'
],inplace=True)
```

```python
final_df= final_df[final_df['info.gender'] == 'male']
final_df.drop(columns=['info.gender'],inplace = True)
```

```python
final_df = final_df[final_df['info.overs'] == 20]
final_df.drop(columns=['info.overs','info.match_type'],inplace=True)
```

```python
matches = final_df
```

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    frames = [delivery_df, loop_df]
    delivery_df = pd.concat(frames)
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```

```python
out = delivery_df[['match_id','batting_team','bowling_team','ball','runs','player_dismissed','city','venue']]
```

```python
out
```

```python
c = np.where(out['city'].isnull(),out['venue'].str.split().apply(lambda x:x[0]),out['city'])
```

```python
out['city'] = c
```

```python
out.drop(columns=['venue'],inplace = True )
```

```python
out
```

```python
out.isnull().sum()
```

```python
eligible = out['city'].value_counts()[out['city'].value_counts()>600].index.tolist()
```

```python
out = out[out['city'].isin(eligible)]
```

```python
out['runs'].dtype
```

```python
total_df = out.groupby('match_id').sum()['runs'].reset_index()
out = out.merge(total_df,on='match_id')
```

```python
out
```

```python
out['current_score'] = out.groupby(['match_id'])['runs_x'].cumsum()
```

```python
out
```

```python
out['over_no'] = out['ball'].apply(lambda x:str(x).split(".")[0])
out['ball_no'] = out['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
out
```

```python
out['balls_bowled'] = (out['over_no'].astype('int')*6) + out['ball_no'].astype('int')
```

```python
out
```

```python
out['balls_left'] = 120 - out['balls_bowled']
out['balls_left'] = out['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
out['player_dismissed'] = out['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
out['player_dismissed'] = out['player_dismissed'].astype('int')
out['player_dismissed'] = out.groupby(['match_id'])['player_dismissed'].cumsum()
out['wickets_left'] = 10 - out['player_dismissed']
```

```python
out['crr'] = round((out['current_score']*6)/out['balls_bowled'],2)
```

```python
out
```

```python
groups = out.groupby('match_id')
match_ids = out['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs_x'].sum().values.tolist())
```

```python
out['last_five'] = last_five
```

```python
out
```

```python
final = out[['match_id','batting_team','bowling_team','runs_x','current_score','balls_bowled','wickets_left','crr','city','runs_y','balls_left','last_five']]
```

```python
final
```

```python
final.drop(columns=['match_id','balls_bowled','runs_x'],inplace = True)
```

```python
final.dropna(inplace = True)
```

```python
final
```

```python
final = final.sample(final.shape[0])
```

```python
X = final.drop(columns=['runs_y'])
y = final['runs_y']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
from xgboost import XGBRegressor
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
